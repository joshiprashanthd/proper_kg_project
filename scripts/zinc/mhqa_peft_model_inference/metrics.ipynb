```python
import pandas as pd
from pathlib import Path
```

```python
folder = Path("mhqa-lora-40-results")
csv_paths = [file for file in folder.iterdir()]
csv_paths[:4]
```

```python
df = None
for path in csv_paths:
    if df is None:
        df = pd.read_csv(path)
    else:
        df = pd.concat([df, pd.read_csv(path)])

df.reset_index(inplace=True, drop=True)
df.info()
```

```python
df[df['correct_option'].isna()]
```

```python
indc = df[df['correct_option'].isna()].index
print(indc)
```

```python
for idx in indc:
    print("INDEX = ", idx)
    cop = int(df.loc[idx, 'raw_response'].split("\nJustification:")[0])
    print("COP = ", cop)
    df.loc[idx, 'correct_option'] = cop
df.info()
```

```python
df.info()
```

```python
df.to_csv("results-mhqa-lora-40.csv", index=False)
```
